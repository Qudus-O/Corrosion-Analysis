# Metal rod corrosion rate with and without protective coating

The dataset can be downloaded from kaggle. click [here](https://www.kaggle.com/datasets/bobinsajigeorge/corrosion-rate) to have access to the dataset.
## About the dataset:
Metal rod is coated with copper, nickel, chromium coatings using electroplating technique and compared against bare metal for corrosion rate.

The tank was filled with 2.0 litres of substitute sea water. The bare specimen and coated specimen of diameter 22mm and length 100mm respectively were kept in the same tank and left for 60 days. Corrosion is noted at certain intervals and photograph was taken.

##  Data Exploration


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('corrosion rate.xlsx')
data
```

## Data cleaning
Some data cleaning processes need to be performed:
+ cleaning the header row
+ deleting the empty columns
+ rename the column name

```python
del data['Unnamed: 0']
data.columns = data.iloc[0]
cleaned_data = data[1:]
cleaned_data.rename(columns={'Weight of Bare Sepcimen (g)':'Weight of Bare Specimen (g)'}, inplace= True)
cleaned_data
```

## Data visualization 

Using matplot library to visualize our analysis.

plotting the graph of time against each of the weights of the specimen.

```python
plt.plot(cleaned_data['Days'], cleaned_data['Weight of Bare Specimen (g)'],label= 'Bare_Metal')
plt.plot(cleaned_data['Days'], cleaned_data['Weight of Coated Specimen (g)'], label= 'Coated_Metal')
plt.title('Corrosion rate of metal rod with and without protective coating')
plt.xlabel('Time in Days')
plt.ylabel('Weight of Specimen (g)')
plt.legend()
plt.show()
```

## Conclusion
+ The analyis indicated that coated metals are less corrosive while bare metals corrode rapidly.

+ At the end of the experiment(day 60), the bare metal lose 33% of it material, while the coated metal loses 3.7% of its material.

Note: Protective Coating is one of the ways to prevent metal from corrosion. some other means are:
